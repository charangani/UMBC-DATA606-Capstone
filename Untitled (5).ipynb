# Predictive Models

#### Data Split

```python
data.columns
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.drop('Exited', axis=1), data['Exited'], test_size=0.25,shuffle=True)
```

#### Standardization and Model Training

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# Standardizing the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Training the logistic regression model
model = LogisticRegression()
model.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report
# Predicting on the test set
y_pred = model.predict(X_test_scaled)

# Calculating the accuracy for the test set
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
report = classification_report(y_test, y_pred)
print(report)
```

The logistic model is quite good at predicting customers who will not exit, but it is struggling to correctly identify a significant portion of customers who will exit. 

### Implementing Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
# Training the Decision Tree classifier
model = DecisionTreeClassifier(max_depth=10)
model.fit(X_train_scaled, y_train)
```

```python
# Making predictions on the test set
y_pred = model.predict(X_test_scaled)
```

```python
report = classification_report(y_test, y_pred)
print(report)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Plotting the decision tree
plt.figure(figsize=(20,10))
plot_tree(model, filled=True)
plt.show()
```

### Implementing Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(max_depth=15, n_estimators=20)

rf_model.fit(X_train_scaled, y_train)
```

```python
# Making predictions
y_pred = rf_model.predict(X_test_scaled)

# Evaluating the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
print(classification_report(y_test, y_pred))
```
